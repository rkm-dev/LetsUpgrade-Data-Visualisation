### Question 1 -> Plot using sns.relplot, line plot, box plot on fmri data set

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
frmi = sns.load_dataset('fmri')
frmi.head()
```

```python
sns.set(style='darkgrid')
sns.set(rc={'figure.figsize':(15,7)})

sns.relplot(x='timepoint', y='signal', data=frmi,  hue='region');
```

```python
sns.relplot(x='subject', y='signal', data=frmi, hue='timepoint', style='region');
```

```python
sns.lineplot(x='timepoint', y='signal', data=frmi, hue='region', style='event');
```

```python
sns.boxplot(x='subject', y='signal', data=frmi, hue='region');
```
