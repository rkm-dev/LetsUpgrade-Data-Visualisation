### Question 1 -> plot a line plot between a and b:
```
a = np.arange(40,50)
b = np.arange(50,60)
```

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
a = np.arange(40,50)
b = np.arange(50,60)

plt.plot(a,b)
plt.title("Line plot")
plt.xlabel("x-axis a")
plt.ylabel("y-axis b")
````

### Question 2 -> Plot a line plot showing the sales trend in company 1 and 2:
```
days    = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
```

````python
days    = [1,2,3,4,5,6,7]                #days of d week
sales_1 = [160,150,140,145,175,165,180]  #sales of company1
sales_2 = [70,90,160,150,140,145,175]    #sales of company2

line1, = plt.plot(days, sales_1, color="green", label="sales of company1")
line2, = plt.plot(days, sales_2, color="blue", label="sales of company2")
plt.legend(handles = [line1, line2])
````

### Question 3 -> Create a 3 by 3 subplots:

multiple plots

````python
x  = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]

f = plt.figure()
f.set_figwidth(15)
f.set_figheight(10)

plt.subplot(3,3,1)
plt.plot(x,y1, "g--")

plt.subplot(3,3,2)
plt.plot(x,y2, "b")

plt.subplot(3,3,3)
plt.plot(x,y3, "h-")

plt.subplot(3,3,4)
plt.plot(x,y4, "yd-")

plt.subplot(3,3,5)
plt.plot(x,y5, "m*-");
````
